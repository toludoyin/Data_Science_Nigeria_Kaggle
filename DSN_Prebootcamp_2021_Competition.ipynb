## Splitting Data

```python
from sklearn.model_selection import train_test_split
feature_col_names=['Earnings', 'Customer_Worth', 'Premium_Auto_per_Month','Total Claim Amount', 'Months Since Last Claim',
'Months Since Policy Inception', 'Counts_of_Open_Complaints','No_Of_POlicy', 'Sex_Female', 'Sex_Male',
'Region_Code_Rural', 'Region_Code_Suburban', 'Region_Code_Urban','Marital Status_Divorced', 'Marital Status_Married',
'Marital Status_Single', 'State_of_Origin_Abuja','State_of_Origin_Enugu', 'State_of_Origin_Imo', 'State_of_Origin_Kano',
'State_of_Origin_Lagos', 'Customer_Reply_Interested','Customer_Reply_Not_interested', 'Plan_Bronze', 'Plan_Gold',
'Plan_Silver', 'Certification_First_Degree', 'Certification_HND','Certification_O_Level', 'Certification_PHD',
'Certification_Second_Degree', 'Status_of_Employment_Disabled','Status_of_Employment_Fully_Employed','Status_of_Employment_In_Retirement',
'Status_of_Employment_Not_employed', 'Status_of_Employment_On_Leave','Vehicle Class_Four-Door Car', 'Vehicle Class_Luxury Car',
'Vehicle Class_Luxury SUV', 'Vehicle Class_SUV','Vehicle Class_Sports Car', 'Vehicle Class_Two-Door Car',
'Vehicle Size_Large', 'Vehicle Size_Medsize', 'Vehicle Size_Small','kind_of_policy_Corporate', 'kind_of_policy_Personal',
'kind_of_policy_Special', 'Sales Channel_Agent', 'Sales Channel_Branch','Sales Channel_Call Center', 'Sales Channel_Web']
predicted_class_names=['Best_Offer']
X=df_dummies[feature_col_names].values
y=df_dummies[predicted_class_names].values 
split_test_size=0.30
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=split_test_size, random_state=42)
```

```python
print("{0:0.2f}% in training set".format((len(X_train)/len(df_dummies.index)) * 100))
print("{0:0.2f}% in test set".format((len(X_test)/len(df_dummies.index)) * 100))
```

## Train Data

```python
from sklearn.ensemble import GradientBoostingRegressor
gbr=GradientBoostingRegressor()
gbr.fit(X_train, y_train.ravel())
```

```python
gbrpred=gbr.predict(X_train)
gbrpred_test=gbr.predict(X_test)
```

```python
MSE = mean_squared_error(y_train, gbrpred)
MSE_test= mean_squared_error(y_test, gbrpred_test)
print("MSE:", MSE)
print("MSE Test:", MSE_test)
RMSE=sqrt(MSE)
RMSE_test=sqrt(MSE_test)
print("RMSE:", RMSE)
print("RMSE Test:", RMSE_test)
```
